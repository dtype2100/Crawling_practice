# 환경설정

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

# Crawler 

```python
title = []
start_num = 1
end_num = 41

while start_num <= end_num:
  url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%8F%84%EC%8B%9C&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=88&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=" + str(start_num)
  reponse = requests.get(url)
  soup = bs(reponse.text, 'html.parser')
  title_tags = soup.select('.news_tit')
  start_num += 10
  print(str(start_num)+"번째 완료")
  time.sleep(2)
  for tag in title_tags:
    title.append(tag.text)
```

# DataFrame 만들기

```python
df = pd.DataFrame(title, columns=['title'])
```
